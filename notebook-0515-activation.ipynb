## 活性化関数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_function(func):
    x = np.linspace(-6, 6, 100)
    y = func(x)

    plt.plot(x, y)
    plt.grid()
    plt.show()
```

## step

```python
def step(x):
    return np.array(x > 0, dtype=np.int)

plot_function(step)
```

## relu

```python
def relu(x):
    return np.maximum(x, 0)

plot_function(relu)
```

## sigmoid

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

plot_function(sigmoid)
```

## softmax

```python
def softmax(x):
    x = x - np.max(x)
    return np.exp(x) / np.sum(np.exp(x))

plot_function(softmax)
```

## tanh

```python
def tanh(x):
    return np.tanh(x)

plot_function(tanh)
```
