```python
import numpy as np
import matplotlib.pyplot as plt
```

# OR

```python
x = np.array([1, 0])
w = np.array([0.5, 0.5])
b = -0.2
hx = np.dot(x, w) + b
y = 1 if hx > 0 else 0
print(y)
```

```python
x1 = np.arange(-2, 2, 0.1)
x2 = (-0.5*x1 + 0.2) / 0.5

plt.plot(x1, x2)
plt.scatter([0, 1, 1], [1, 1, 0], c="r")
plt.scatter(0, 0, c="b")
plt.grid()
plt.show()
```

## AND

```python
x = np.array([1, 1])
w = np.array([0.5, 0.5])
b = -0.7
hx = np.dot(x, w) + b
y = 1 if hx > 0 else 0
print(y)
```

```python
x1 = np.arange(-2, 2, 0.1)
x2 = (-0.5*x1 + 0.7) / 0.5

plt.plot(x1, x2)
plt.scatter(1, 1, c="r")
plt.scatter([0, 0, 1], [0, 1, 0], c="b")
plt.grid()
plt.show()
```

## NAND

```python
x = np.array([0, 1])
w = np.array([-0.5, -0.5])
b = 0.7
hx = np.dot(x, w) + b
y = 1 if hx > 0 else 0
print(y)
```

```python
x1 = np.arange(-2, 2, 0.1)
x2 = -(0.5*x1 - 0.7) / 0.5

plt.plot(x1, x2)
plt.scatter([0, 0, 1], [0, 1, 0], c="r")
plt.scatter(1, 1, c="b")
plt.grid()
plt.show()
```

## XOR

```python
# 活性化関数を定義
def step_function(x):
    return np.array(x > 0, dtype = np.int)

# 入力
x_in = np.array([0, 1])

# 重みとバイアスの初期化
w1 = np.array([[-0.5, -0.5],
               [0.5, 0.5]])
b1 = np.array([0.7, -0.2])
w2 = np.array([0.5, 0.5])
b2 = -0.7

# 1層目
a1 = np.dot(x_in, w1.T) + b1
# 活性化関数
z1 = step_function(a1)
# 2層目
a2 = np.dot(z1, w2) + b2
# 活性化関数
z2 = step_function(a2)

print(z2)
```
