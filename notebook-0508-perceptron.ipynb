```python
import numpy as np
import matplotlib.pyplot as plt
```

# OR

```python
x = np.array([1, 0])
w = np.array([0.5, 0.5])
b = -0.2
hx = np.dot(x, w) + b
y = 1 if hx >= 0 else 0
print(y)
```

```python
x1 = np.arange(-2, 2, 0.1)
x2 = (-0.5*x1 + 0.2) / 0.5

plt.plot(x1, x2)
plt.scatter([0, 1, 1], [1, 1, 0], c="r")
plt.scatter(0, 0, c="b")
plt.grid()
plt.show()
```

## AND

```python
x = np.array([1, 1])
w = np.array([0.5, 0.5])
b = -0.7
hx = np.dot(x, w) + b
y = 1 if hx >= 0 else 0
print(y)
```

```python
x1 = np.arange(-2, 2, 0.1)
x2 = (-0.5*x1 + 0.7) / 0.5

plt.plot(x1, x2)
plt.scatter(1, 1, c="r")
plt.scatter([0, 0, 1], [0, 1, 0], c="b")
plt.grid()
plt.show()
```
